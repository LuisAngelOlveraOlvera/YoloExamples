```python
import torch
import zipfile
import os

# Descargar el archivo zip
url = 'https://ultralytics.com/assets/coco2017val.zip'
zip_file = 'tmp.zip'
dest_dir = r'C:\Users\User\Pictures\dataset'  # Ruta de destino

torch.hub.download_url_to_file(url, zip_file)

# Verificar si el archivo se ha descargado correctamente
if os.path.exists(zip_file):
    # Crear el directorio de destino si no existe
    os.makedirs(dest_dir, exist_ok=True)
    
    # Descomprimir el archivo
    with zipfile.ZipFile(zip_file, 'r') as zip_ref:
        zip_ref.extractall(dest_dir)
    
    # Eliminar el archivo zip después de descomprimir
    os.remove(zip_file)
    print(f'El archivo {zip_file} ha sido descomprimido y eliminado.')
else:
    print(f'El archivo {zip_file} no se ha descargado correctamente.')

```

```python
from ultralytics import YOLO
import cv2
import torch
from ultralytics import RTDETR

# Verificar si hay GPU disponible
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

# Cargar el modelo pesado (YOLOv8x)
# model = YOLO("yolov8s-seg.pt").to(device)
model = RTDETR("rtdetr-l.pt").to(device)

# Capturar video desde la cámara
cap = cv2.VideoCapture(0)

# Configuración para procesamiento en lotes y reducción de resolución
batch_size = 4  # Tamaño del lote para procesar varios cuadros a la vez
frame_skip = 3  # Procesar 1 de cada 3 cuadros (opcional para ahorrar procesamiento)
frames = []
count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Saltar cuadros para ahorrar procesamiento
    if count % frame_skip == 0:
        # Redimensionar el frame para mejorar la velocidad
        frame_resized = cv2.resize(frame, (640, 480))  # Ajusta la resolución según tu preferencia

        # Añadir el frame redimensionado al lote de frames
        frames.append(frame_resized)

        # Procesar cuando se alcanza el tamaño del lote
        if len(frames) == batch_size:
            # Realizar la predicción con precisión flotante de 16 bits (FP16)
            results = model.predict(source=frames, half=True)  # Realizar predicción sin mostrar todas las ventanas

            for i, result in enumerate(results):
                # Obtener el frame con las anotaciones de los resultados
                annotated_frame = result.plot()

                # Mostrar el frame anotado en una única ventana
                cv2.imshow('Detecciones', annotated_frame)

            frames = []  # Limpiar el lote después de la predicción

            # Liberar memoria de la GPU para evitar cuellos de botella
            torch.cuda.empty_cache()

    count += 1

    # Presionar 'q' para salir del bucle
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Liberar recursos
cap.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import RTDETR
import cv2
import torch
import time

# Verificar si hay GPU disponible
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

# Cargar el modelo
model = RTDETR("rtdetr-l.pt").to(device)

# Capturar video desde la cámara
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)

# Configuración para procesamiento
frame_skip = 2  # Procesar 1 de cada 2 cuadros
count = 0

# Configurar CUDA para rendimiento
torch.backends.cudnn.benchmark = True
torch.backends.cuda.matmul.allow_tf32 = True

while True:
    ret, frame = cap.read()
    if not ret:
        break

    if count % frame_skip == 0:
        # Redimensionar el frame
        frame_resized = cv2.resize(frame, (640, 480))
        
        # Realizar la predicción con precisión flotante de 16 bits (FP16)
        with torch.cuda.amp.autocast():
            results = model.predict(source=frame_resized, half=True, verbose=False)
        
        # Obtener el frame con las anotaciones de los resultados
        annotated_frame = results[0].plot()
        
        # Mostrar el frame anotado
        cv2.imshow('Detecciones', annotated_frame)
    
    count += 1
    
    # Presionar 'q' para salir del bucle
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Liberar recursos
cap.release()
cv2.destroyAllWindows()
```
