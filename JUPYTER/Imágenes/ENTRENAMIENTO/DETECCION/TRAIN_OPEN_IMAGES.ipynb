```python
import os
import glob
from PIL import Image
import yaml
import numpy as np

def validate_dataset(dataset_path):
    print("Iniciando validación del dataset...")
    
    # 1. Verificar estructura de directorios
    required_dirs = ['images/train', 'images/val', 'labels/train', 'labels/val']
    for dir_path in required_dirs:
        full_path = os.path.join(dataset_path, dir_path)
        if not os.path.exists(full_path):
            print(f"❌ Error: No se encuentra el directorio: {full_path}")
            return False
        else:
            print(f"✅ Directorio encontrado: {dir_path}")
    
    # 2. Verificar correspondencia entre imágenes y etiquetas
    for split in ['train', 'val']:
        images_path = os.path.join(dataset_path, 'images', split)
        labels_path = os.path.join(dataset_path, 'labels', split)
        
        image_files = set([os.path.splitext(os.path.basename(f))[0] 
                         for f in glob.glob(os.path.join(images_path, '*.*'))
                         if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp'))])
        
        label_files = set([os.path.splitext(os.path.basename(f))[0] 
                         for f in glob.glob(os.path.join(labels_path, '*.txt'))])
        
        print(f"\nAnalizando conjunto {split}:")
        print(f"Imágenes encontradas: {len(image_files)}")
        print(f"Etiquetas encontradas: {len(label_files)}")
        
        # Verificar imágenes sin etiquetas
        images_without_labels = image_files - label_files
        if images_without_labels:
            print(f"❌ Imágenes sin etiquetas ({len(images_without_labels)}):")
            for img in list(images_without_labels)[:5]:  # Mostrar solo los primeros 5
                print(f"  - {img}")
            if len(images_without_labels) > 5:
                print("  ...")
        
        # Verificar etiquetas sin imágenes
        labels_without_images = label_files - image_files
        if labels_without_images:
            print(f"❌ Etiquetas sin imágenes ({len(labels_without_images)}):")
            for lbl in list(labels_without_images)[:5]:  # Mostrar solo los primeros 5
                print(f"  - {lbl}")
            if len(labels_without_images) > 5:
                print("  ...")
    
    # 3. Verificar formato de etiquetas y valores
    print("\nVerificando formato de etiquetas...")
    label_errors = []
    for split in ['train', 'val']:
        labels_path = os.path.join(dataset_path, 'labels', split)
        for label_file in glob.glob(os.path.join(labels_path, '*.txt')):
            try:
                with open(label_file, 'r') as f:
                    lines = f.readlines()
                for line_num, line in enumerate(lines, 1):
                    try:
                        values = [float(x) for x in line.strip().split()]
                        if len(values) != 5:
                            label_errors.append(f"{label_file} (línea {line_num}): Número incorrecto de valores")
                        else:
                            class_id, x, y, w, h = values
                            if not (0 <= x <= 1 and 0 <= y <= 1 and 0 <= w <= 1 and 0 <= h <= 1):
                                label_errors.append(f"{label_file} (línea {line_num}): Valores fuera de rango [0,1]")
                            if int(class_id) not in range(8):  # Asumiendo 8 clases (0-7)
                                label_errors.append(f"{label_file} (línea {line_num}): ID de clase inválido")
                    except ValueError:
                        label_errors.append(f"{label_file} (línea {line_num}): Error en formato de valores")
            except Exception as e:
                label_errors.append(f"{label_file}: Error al leer archivo - {str(e)}")
    
    if label_errors:
        print("\n❌ Errores encontrados en etiquetas:")
        for error in label_errors[:10]:  # Mostrar solo los primeros 10 errores
            print(f"  - {error}")
        if len(label_errors) > 10:
            print(f"  ... y {len(label_errors) - 10} errores más")
    else:
        print("✅ Formato de etiquetas correcto")
    
    # 4. Verificar integridad de imágenes
    print("\nVerificando integridad de imágenes...")
    corrupt_images = []
    for split in ['train', 'val']:
        images_path = os.path.join(dataset_path, 'images', split)
        for img_file in glob.glob(os.path.join(images_path, '*.*')):
            if img_file.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')):
                try:
                    with Image.open(img_file) as img:
                        img.verify()
                except Exception as e:
                    corrupt_images.append(f"{img_file}: {str(e)}")
    
    if corrupt_images:
        print("\n❌ Imágenes corruptas encontradas:")
        for error in corrupt_images[:10]:
            print(f"  - {error}")
        if len(corrupt_images) > 10:
            print(f"  ... y {len(corrupt_images) - 10} imágenes corruptas más")
    else:
        print("✅ Todas las imágenes están íntegras")

    return not (label_errors or corrupt_images or images_without_labels or labels_without_images)

# Ruta al dataset
dataset_path = r"C:\Users\User\Documents\JUPYTER_LOCAL\IMAGENES_ENTRENAMIENTO\DETECCION\OpenImagesV2\dataset"

# Ejecutar validación
is_valid = validate_dataset(dataset_path)
print("\nResumen:")
if is_valid:
    print("✅ Dataset válido y listo para usar")
else:
    print("❌ Se encontraron problemas en el dataset que necesitan ser corregidos")
```

```python
import os
import glob

def examine_and_fix_labels(dataset_path):
    print("Examinando archivos de etiquetas...")
    
    # Leer un archivo de ejemplo para ver el formato actual
    labels_path = os.path.join(dataset_path, 'labels', 'train')
    example_file = os.path.join(labels_path, '0000048549557964.txt')
    
    print("\nMostrando contenido del archivo de ejemplo:")
    with open(example_file, 'r') as f:
        content = f.read()
        print(content)
    
    # Función para verificar y corregir el formato de una línea
    def fix_line(line):
        try:
            # Eliminar espacios extras y caracteres no deseados
            line = line.strip()
            # Separar los valores
            values = line.split()
            
            if len(values) != 5:
                return None
            
            # Convertir y validar valores
            class_id = int(float(values[0]))
            x = float(values[1])
            y = float(values[2])
            w = float(values[3])
            h = float(values[4])
            
            # Validar rangos
            if not (0 <= x <= 1 and 0 <= y <= 1 and 0 <= w <= 1 and 0 <= h <= 1):
                return None
            if not (0 <= class_id <= 7):  # Ajustar según tus clases
                return None
                
            # Devolver línea formateada
            return f"{class_id} {x:.6f} {y:.6f} {w:.6f} {h:.6f}"
        except:
            return None

    print("\nIntentando reparar un archivo de ejemplo...")
    with open(example_file, 'r') as f:
        lines = f.readlines()
        fixed_lines = [fix_line(line) for line in lines]
        fixed_lines = [line for line in fixed_lines if line is not None]
    
    print("\nContenido corregido:")
    for line in fixed_lines:
        print(line)
    
    response = input("\n¿El formato corregido parece correcto? (s/n): ")
    if response.lower() != 's':
        print("Abortando el proceso de corrección")
        return

    print("\nIniciando corrección de todos los archivos...")
    error_files = []
    fixed_files = 0
    empty_files = 0
    
    for split in ['train', 'val']:
        labels_path = os.path.join(dataset_path, 'labels', split)
        for label_file in glob.glob(os.path.join(labels_path, '*.txt')):
            try:
                with open(label_file, 'r') as f:
                    lines = f.readlines()
                
                fixed_lines = [fix_line(line) for line in lines]
                fixed_lines = [line for line in fixed_lines if line is not None]
                
                if not fixed_lines:
                    empty_files += 1
                    error_files.append(f"{label_file}: Sin líneas válidas")
                    continue
                
                # Crear backup del archivo original
                backup_file = label_file + '.backup'
                if not os.path.exists(backup_file):
                    os.rename(label_file, backup_file)
                
                # Escribir archivo corregido
                with open(label_file, 'w') as f:
                    for line in fixed_lines:
                        f.write(line + '\n')
                
                fixed_files += 1
                
            except Exception as e:
                error_files.append(f"{label_file}: {str(e)}")
    
    print(f"\nProceso completado:")
    print(f"- Archivos corregidos: {fixed_files}")
    print(f"- Archivos vacíos encontrados: {empty_files}")
    print(f"- Archivos con errores: {len(error_files)}")
    
    if error_files:
        print("\nArchivos con errores:")
        for error in error_files[:10]:
            print(f"- {error}")
        if len(error_files) > 10:
            print(f"... y {len(error_files) - 10} errores más")

# Ruta al dataset
dataset_path = r"C:\Users\User\Documents\JUPYTER_LOCAL\IMAGENES_ENTRENAMIENTO\DETECCION\OpenImagesV2\dataset"

# Ejecutar el script
examine_and_fix_labels(dataset_path)
```

```python
# Script 1: Verificar archivos de etiquetas (txt)

import os
import glob
import numpy as np

def verificar_etiquetas(path_labels, num_clases):
    """
    Verifica los archivos de etiquetas:
    1. Formato correcto (5 columnas)
    2. Valores dentro de rango
    3. Clase válida
    4. Correspondencia con imágenes
    """
    errores = []
    archivos_procesados = 0
    
    # Obtener todos los archivos .txt
    archivos_txt = glob.glob(os.path.join(path_labels, '*.txt'))
    
    print(f"Verificando {len(archivos_txt)} archivos de etiquetas en {path_labels}")
    
    for txt_path in archivos_txt:
        archivos_procesados += 1
        nombre_base = os.path.basename(txt_path)
        
        try:
            # Leer el archivo
            with open(txt_path, 'r') as f:
                lineas = f.readlines()
            
            if len(lineas) == 0:
                errores.append(f"Archivo vacío: {nombre_base}")
                continue
                
            for num_linea, linea in enumerate(lineas, 1):
                try:
                    # Verificar formato
                    valores = list(map(float, linea.strip().split()))
                    
                    if len(valores) != 5:
                        errores.append(f"Formato incorrecto en {nombre_base}, línea {num_linea}: {linea.strip()}")
                        continue
                    
                    clase, x_center, y_center, width, height = valores
                    
                    # Verificar clase
                    if not (0 <= clase < num_clases) or not clase.is_integer():
                        errores.append(f"Clase inválida en {nombre_base}, línea {num_linea}: {clase}")
                    
                    # Verificar coordenadas
                    if not all(0 <= val <= 1 for val in [x_center, y_center, width, height]):
                        errores.append(f"Valores fuera de rango en {nombre_base}, línea {num_linea}")
                        
                except ValueError:
                    errores.append(f"Error de conversión en {nombre_base}, línea {num_linea}: {linea.strip()}")
                    
        except Exception as e:
            errores.append(f"Error al procesar {nombre_base}: {str(e)}")
        
        # Mostrar progreso
        if archivos_procesados % 1000 == 0:
            print(f"Procesados {archivos_procesados}/{len(archivos_txt)} archivos...")
    
    return errores
```

```python
# Script 2: Verificar imágenes

import os
import cv2
import glob
from PIL import Image
import numpy as np
from concurrent.futures import ThreadPoolExecutor
from tqdm import tqdm

def verificar_imagen(img_path):
    """
    Verifica si una imagen es legible y válida
    Retorna (path, estado, mensaje)
    """
    try:
        # Intentar abrir con PIL
        img = Image.open(img_path)
        img.verify()
        
        # Intentar abrir con OpenCV
        img_cv = cv2.imread(img_path)
        if img_cv is None:
            return img_path, False, "No se puede leer con OpenCV"
        
        # Verificar dimensiones
        if img_cv.shape[0] < 10 or img_cv.shape[1] < 10:
            return img_path, False, f"Imagen demasiado pequeña: {img_cv.shape}"
        
        # Verificar si hay contenido
        if img_cv.size == 0 or np.all(img_cv == 0):
            return img_path, False, "Imagen vacía"
            
        return img_path, True, "OK"
        
    except Exception as e:
        return img_path, False, str(e)

def verificar_imagenes(path_imagenes, extensiones=('.jpg', '.jpeg', '.png')):
    """
    Verifica todas las imágenes en el directorio
    """
    errores = []
    imagenes_ok = []
    
    # Obtener lista de imágenes
    archivos = []
    for ext in extensiones:
        archivos.extend(glob.glob(os.path.join(path_imagenes, f'*{ext}')))
    
    print(f"Verificando {len(archivos)} imágenes en {path_imagenes}")
    
    # Verificar imágenes en paralelo
    with ThreadPoolExecutor(max_workers=8) as executor:
        resultados = list(tqdm(executor.map(verificar_imagen, archivos), total=len(archivos)))
    
    # Procesar resultados
    for path, estado, mensaje in resultados:
        if not estado:
            errores.append(f"Error en {os.path.basename(path)}: {mensaje}")
        else:
            imagenes_ok.append(path)
    
    return imagenes_ok, errores
```

```python
import os
import glob
import numpy as np
from collections import Counter

def analizar_errores_etiquetas(path_labels, num_clases):
    """
    Análisis detallado de los errores en las etiquetas
    """
    errores_por_tipo = Counter()
    detalles_errores = []
    estadisticas = {
        'total_archivos': 0,
        'total_lineas': 0,
        'clases_encontradas': set(),
        'min_coords': [float('inf')] * 4,
        'max_coords': [float('-inf')] * 4
    }
    
    archivos_txt = glob.glob(os.path.join(path_labels, '*.txt'))
    print(f"Analizando {len(archivos_txt)} archivos...")
    
    for txt_path in archivos_txt:
        estadisticas['total_archivos'] += 1
        nombre_base = os.path.basename(txt_path)
        
        try:
            with open(txt_path, 'r') as f:
                lineas = f.readlines()
                
            estadisticas['total_lineas'] += len(lineas)
            
            for num_linea, linea in enumerate(lineas, 1):
                try:
                    valores = list(map(float, linea.strip().split()))
                    
                    # Error de formato
                    if len(valores) != 5:
                        errores_por_tipo['formato_incorrecto'] += 1
                        detalles_errores.append(f"Formato ({len(valores)} valores): {nombre_base}:{num_linea}")
                        continue
                    
                    clase, x, y, w, h = valores
                    
                    # Actualizar estadísticas
                    estadisticas['clases_encontradas'].add(int(clase))
                    for i, val in enumerate([x, y, w, h]):
                        estadisticas['min_coords'][i] = min(estadisticas['min_coords'][i], val)
                        estadisticas['max_coords'][i] = max(estadisticas['max_coords'][i], val)
                    
                    # Verificar clase
                    if not clase.is_integer():
                        errores_por_tipo['clase_decimal'] += 1
                        detalles_errores.append(f"Clase decimal {clase}: {nombre_base}:{num_linea}")
                    if not (0 <= clase < num_clases):
                        errores_por_tipo['clase_invalida'] += 1
                        detalles_errores.append(f"Clase fuera de rango {clase}: {nombre_base}:{num_linea}")
                    
                    # Verificar coordenadas
                    for val, nombre in zip([x, y, w, h], ['x', 'y', 'w', 'h']):
                        if not (0 <= val <= 1):
                            errores_por_tipo[f'coordenada_{nombre}_fuera_rango'] += 1
                            detalles_errores.append(f"{nombre}={val}: {nombre_base}:{num_linea}")
                    
                except ValueError:
                    errores_por_tipo['error_conversion'] += 1
                    detalles_errores.append(f"Error conversión: {nombre_base}:{num_linea}")
                    
        except Exception as e:
            errores_por_tipo['error_archivo'] += 1
            detalles_errores.append(f"Error archivo {nombre_base}: {str(e)}")
    
    return errores_por_tipo, detalles_errores, estadisticas

# Ejecutar análisis
if __name__ == "__main__":
    PATH_DATASET = r"C:\Users\User\Documents\JUPYTER_LOCAL\IMAGENES_ENTRENAMIENTO\DETECCION\OpenImagesV2\dataset"
    NUM_CLASES = 8
    
    for conjunto in ['train', 'val']:
        print(f"\nAnalizando etiquetas de {conjunto}:")
        path_labels = os.path.join(PATH_DATASET, 'labels', conjunto)
        
        errores, detalles, stats = analizar_errores_etiquetas(path_labels, NUM_CLASES)
        
        print("\nEstadísticas generales:")
        print(f"Total archivos: {stats['total_archivos']}")
        print(f"Total líneas: {stats['total_lineas']}")
        print(f"Clases encontradas: {sorted(stats['clases_encontradas'])}")
        print(f"Rango coordenadas X: [{stats['min_coords'][0]:.3f}, {stats['max_coords'][0]:.3f}]")
        print(f"Rango coordenadas Y: [{stats['min_coords'][1]:.3f}, {stats['max_coords'][1]:.3f}]")
        print(f"Rango ancho: [{stats['min_coords'][2]:.3f}, {stats['max_coords'][2]:.3f}]")
        print(f"Rango alto: [{stats['min_coords'][3]:.3f}, {stats['max_coords'][3]:.3f}]")
        
        print("\nResumen de errores:")
        for tipo, cantidad in errores.most_common():
            print(f"{tipo}: {cantidad}")
        
        # Guardar detalles en archivo
        with open(f'analisis_detallado_{conjunto}.txt', 'w') as f:
            f.write("\n".join(detalles))
        
        print(f"\nDetalles guardados en analisis_detallado_{conjunto}.txt")
```

```python
import os

# Diccionario de clases: mapea los nombres de clase a índices
class_mapping = {
    "cat": "0",
    "human_arm": "1",
    "car": "2",
    "human_eye": "3",
    "human_face": "4",
    "human_hand": "5",
    "human_mouth": "6",
    "person": "7"
}

# Ruta a las carpetas de train y val
train_dir = r"C:\Users\User\Documents\JUPYTER_LOCAL\IMAGENES_ENTRENAMIENTO\DETECCION\OpenImagesV2\dataset\labels\train"
val_dir = r"C:\Users\User\Documents\JUPYTER_LOCAL\IMAGENES_ENTRENAMIENTO\DETECCION\OpenImagesV2\dataset\labels\val"

def update_labels(directory):
    # Itera sobre cada archivo .txt en el directorio
    for filename in os.listdir(directory):
        if filename.endswith(".txt"):
            file_path = os.path.join(directory, filename)
            
            # Lee el contenido del archivo
            with open(file_path, "r") as file:
                lines = file.readlines()
            
            # Actualiza las etiquetas según el mapeo
            new_lines = []
            for line in lines:
                parts = line.strip().split()
                if parts and parts[0] in class_mapping:
                    # Reemplaza la etiqueta textual por el índice correspondiente
                    parts[0] = class_mapping[parts[0]]
                    new_lines.append(" ".join(parts))
            
            # Escribe el nuevo contenido en el archivo
            with open(file_path, "w") as file:
                file.write("\n".join(new_lines))

# Actualiza las etiquetas en las carpetas de train y val
update_labels(train_dir)
update_labels(val_dir)

print("Actualización completada.")
```
