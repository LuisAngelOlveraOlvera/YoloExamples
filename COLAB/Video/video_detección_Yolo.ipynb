```python
from ultralytics import YOLO
import os

# Load the pre-trained YOLOv8 model
model = YOLO('yolov8s.pt')

# Create a directory to save the output video
save_dir = "/content/YOLO-Pre-Trained-Results"
os.makedirs(save_dir, exist_ok=True)

# Perform inference on the video with stream=True to avoid memory issues
results = model.predict('gatos.mp4', stream=True, save=True, project=save_dir, name="salida")

# Process the results
for r in results:
    boxes = r.boxes  # Boxes object for bbox outputs
    masks = r.masks  # Masks object for segment masks outputs
    probs = r.probs  # Class probabilities for classification outputs

# Verify if the video has been saved correctly
if os.path.exists(save_dir):
    print(f"The video has been saved at: {save_dir}")
else:
    print("The video could not be saved.")
```

```python
import cv2
import numpy as np
import base64
from IPython.display import HTML
from IPython.display import display
import time

def play_video(video_path, fps=30):
    cap = cv2.VideoCapture(video_path)

    html = f'''
    <div style="display: flex; justify-content: center;">
        <img id="video" style="max-width: 100%; height: auto;">
    </div>
    <script>
        var img = document.getElementById("video");
        var frames = [];
        var currentFrame = 0;
        var fps = {fps};

        function updateImage() {{
            if (currentFrame < frames.length) {{
                img.src = frames[currentFrame];
                currentFrame++;
                setTimeout(updateImage, 1000 / fps);
            }}
        }}

        function addFrame(frame) {{
            frames.push(frame);
            if (frames.length === 1) {{
                updateImage();
            }}
        }}
    </script>
    '''

    display(HTML(html))

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        _, buffer = cv2.imencode('.jpg', frame)
        jpg_as_text = base64.b64encode(buffer).decode()
        display(HTML(f'<script>addFrame("data:image/jpeg;base64,{jpg_as_text}")</script>'))

        time.sleep(0.05)  # Aumentar el intervalo para reducir la carga en la CPU

    cap.release()

# Uso de la funci√≥n
video_path = '/content/YOLO-Pre-Trained-Results/salida/gatos.avi'  # Ajusta esta ruta a la ubicaci√≥n de tu video
```
