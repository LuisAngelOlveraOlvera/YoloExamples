```python
import cv2
import numpy as np
import base64
from IPython.display import HTML
from IPython.display import display
import time

def play_video(video_path, fps=30):
    cap = cv2.VideoCapture(video_path)

    html = f'''
    <div style="display: flex; justify-content: center;">
        <img id="video" style="max-width: 100%; height: auto;">
    </div>
    <script>
        var img = document.getElementById("video");
        var frames = [];
        var currentFrame = 0;
        var fps = {fps};

        function updateImage() {{
            if (currentFrame < frames.length) {{
                img.src = frames[currentFrame];
                currentFrame++;
                setTimeout(updateImage, 1000 / fps);
            }}
        }}

        function addFrame(frame) {{
            frames.push(frame);
            if (frames.length === 1) {{
                updateImage();
            }}
        }}
    </script>
    '''

    display(HTML(html))

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        _, buffer = cv2.imencode('.jpg', frame)
        jpg_as_text = base64.b64encode(buffer).decode()
        display(HTML(f'<script>addFrame("data:image/jpeg;base64,{jpg_as_text}")</script>'))

        time.sleep(0.05)  # Aumentar el intervalo para reducir la carga en la CPU

    cap.release()

# Uso de la funci√≥n
video_path1 = '/content/YOLO-Pre-Trained-Results/salida/aerial2.avi'  # Ajusta esta ruta a la ubicaci√≥n de tu video
video_path2 = '/content/YOLO-Pre-Trained-Results/salida2/aerial3.avi'  # Ajusta esta ruta a la ubicaci√≥n de tu video
```

```python
play_video(video_path1)
```
