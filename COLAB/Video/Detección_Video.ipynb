```python
from ultralytics import YOLO
import cv2
import os

# Cargar el modelo preentrenado YOLOv8
model = YOLO('yolov8s.pt')

# Crear un directorio donde guardar el video de salida
save_dir = "/content/YOLO-Pre-Trained-Results"
os.makedirs(save_dir, exist_ok=True)

# Definir la ruta del video de salida
output_video_path = os.path.join(save_dir, "salida.mp4")

# Abrir el video de entrada
video = cv2.VideoCapture('gatos.mp4')

# Obtener propiedades del video
width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(video.get(cv2.CAP_PROP_FPS))

# Definir el codec y crear el objeto VideoWriter
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

while True:
    # Leer un frame del video
    ret, frame = video.read()
    if not ret:
        break

    # Realizar la predicci√≥n en el frame
    results = model(frame)

    # Dibujar las detecciones en el frame
    annotated_frame = results[0].plot()

    # Escribir el frame en el video de salida
    out.write(annotated_frame)

# Liberar los objetos de video
video.release()
out.release()

# Verificar si el video ha sido guardado correctamente
if os.path.exists(output_video_path):
    print(f"El video se ha guardado en: {output_video_path}")
else:
    print("No se ha podido guardar el video.")
```

```python
import cv2
import numpy as np
import base64
from IPython.display import HTML
from IPython.display import display
import time

def play_video(video_path, fps=30):
    cap = cv2.VideoCapture(video_path)

    html = f'''
    <div style="display: flex; justify-content: center;">
        <img id="video" style="max-width: 100%; height: auto;">
    </div>
    <script>
        var img = document.getElementById("video");
        var frames = [];
        var currentFrame = 0;
        var fps = {fps};

        function updateImage() {{
            if (currentFrame < frames.length) {{
                img.src = frames[currentFrame];
                currentFrame++;
                setTimeout(updateImage, 1000 / fps);
            }}
        }}

        function addFrame(frame) {{
            frames.push(frame);
            if (frames.length === 1) {{
                updateImage();
            }}
        }}
    </script>
    '''

    display(HTML(html))

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        _, buffer = cv2.imencode('.jpg', frame)
        jpg_as_text = base64.b64encode(buffer).decode()
        display(HTML(f'<script>addFrame("data:image/jpeg;base64,{jpg_as_text}")</script>'))

        time.sleep(0.01)  # Small delay to allow for smooth playback

    cap.release()

# Uso de la funci√≥n
video_path = '/content/YOLO-Pre-Trained-Results/salida/gatos.avi'  # Ajusta esta ruta a la ubicaci√≥n de tu video
play_video(video_path)
```
