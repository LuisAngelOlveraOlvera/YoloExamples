```python
from ultralytics import FastSAM
import os

# Cargar el modelo YOLO m√°s ligero
model = FastSAM("FastSAM-s.pt")  # or FastSAM-x.pt

results = model("/content/minicky_pose.jpg")

save_dir = "/content/YOLO-Pre-Trained-Results"
os.makedirs(save_dir, exist_ok=True)

results = results[0]

file_path = os.path.join(save_dir, "minicky_pose_FAST_SAM.jpg")
results.save(file_path)

# Verificar si la imagen ha sido guardado o no
if os.path.exists(file_path):
    print(f"La imagen se ha guardado en: {file_path}")
else:
    print("No se ha podido guardar la imagen.")
```

```python
import matplotlib.pyplot as plt
import cv2

resultado_segmentacion = "/content/YOLO-Pre-Trained-Results/minicky_pose_FAST_SAM.jpg"
img = cv2.imread(resultado_segmentacion)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
# Primer m√©todo para visualizar im√°genes (sin ciclo for)
import matplotlib.pyplot as plt
import cv2

inf_fast_sam_1 = "/content/YOLO-Pre-Trained-Results/minicky_pose_FAST_SAM.jpg"
inf_sam_1 = "/content/YOLO-Pre-Trained-Results/minicky_pose_SAM.jpg"
inf_seg_1 = "/content/YOLO-Pre-Trained-Results/minicky_pose_SEG.jpg"

img_fast_sam_1 = cv2.imread(inf_fast_sam_1)
img_inf_sam_1 = cv2.imread(inf_sam_1)
img_inf_seg_1 = cv2.imread(inf_seg_1)

fast_sam_1 = cv2.cvtColor(img_fast_sam_1, cv2.COLOR_BGR2RGB)
sam_1 = cv2.cvtColor(img_inf_sam_1, cv2.COLOR_BGR2RGB)
seg_1 = cv2.cvtColor(img_inf_seg_1, cv2.COLOR_BGR2RGB)

fig, axes = plt.subplots(1,3, figsize=(15,5))

axes[0].imshow(fast_sam_1)
axes[0].set_title("Minicky FAST SAM")
axes[0].axis("off")

axes[1].imshow(sam_1)
axes[1].set_title("Minicky SAM")
axes[1].axis("off")


axes[2].imshow(seg_1)
axes[2].set_title("Minicky SEG")
axes[2].axis("off")
```
