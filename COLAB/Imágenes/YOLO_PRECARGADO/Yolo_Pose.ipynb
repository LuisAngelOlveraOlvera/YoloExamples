```python
from ultralytics import YOLO
import os
import matplotlib.pyplot as plt
import cv2

model = YOLO('yolov8s-pose.pt')
results = model('pose1.jpg')

save_dir = save_dir = "/content/YOLO-Pre-Trained-Results"
os.makedirs(save_dir, exist_ok=True)

result = results[0]

file_path = os.path.join(save_dir, 'pose1_result.jpg')
result.save(file_path)

# Verificar si la imagen ha sido guardado o no
if os.path.exists(file_path):
    print(f"La imagen se ha guardado en: {file_path}")
else:
    print("No se ha podido guardar la imagen.")
```

```python
import matplotlib.pyplot as plt
import cv2

pose_1 = "/content/YOLO-Pre-Trained-Results/Resultado_personas5.jpg"
pose_2 = "/content/YOLO-Pre-Trained-Results/Resultado_pose2.jpg"
pose_3 = "/content/YOLO-Pre-Trained-Results/Resultado_pose4.jpg"

img_1 = cv2.imread(pose_1)
img_2 = cv2.imread(pose_2)
img_3 = cv2.imread(pose_3)

img_1_color = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)
img_2_color = cv2.cvtColor(img_2, cv2.COLOR_BGR2RGB)
img_3_color = cv2.cvtColor(img_3, cv2.COLOR_BGR2RGB)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(img_1_color)
axes[0].set_title("Persona")
axes[0].axis("off")

axes[1].imshow(img_2_color)
axes[1].set_title("Pose 2")
axes[1].axis("off")

axes[2].imshow(img_3_color)
axes[2].set_title("Pose 4")
axes[2].axis("off")
```

```python
import matplotlib.pyplot as plt
import cv2
import os

poses = ["/content/YOLO-Pre-Trained-Results/Resultado_personas5.jpg",
         "/content/YOLO-Pre-Trained-Results/Resultado_pose2.jpg",
         "/content/YOLO-Pre-Trained-Results/Resultado_pose4.jpg",
         "/content/YOLO-Pre-Trained-Results/pose1_result.jpg",]

for img_path in poses:
  img = cv2.imread(img_path)
  img_color = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  plt.imshow(img_color)
  plt.title(os.path.basename(img_path))
  plt.axis('off')
  plt.show()
```
