```python
from ultralytics import YOLO
import os

model = YOLO("yolov8s-cls.pt")

results = model("Pelusa.jpeg")

save_dir = "/content/YOLO-Pre-Trained-Results"
os.makedirs(save_dir, exist_ok=True)

results = results[0]

file_path = os.path.join(save_dir, "pelusa_clas.jpg")
results.save(file_path)

# Verificar si la imagen ha sido guardado o no
if os.path.exists(file_path):
    print(f"La imagen se ha guardado en: {file_path}")
else:
    print("No se ha podido guardar la imagen.")
```

```python
from ultralytics import YOLO
import os

modelo = YOLO("yolov8s-cls.pt")

imagenes = ["personas1.png","personas5.jpg", "pose4.jpg", "bus.jpg"]

save_dir = "/content/YOLO-Pre-Trained-Results"
os.makedirs(save_dir, exist_ok=True)

for img_path in imagenes:
  results=modelo(img_path)
  result = results[0]
  output_image_name = f"Resultado_{os.path.basename(img_path)}"
  output_image_path = os.path.join(save_dir, output_image_name)
  result.save(output_image_path)


  if os.path.exists(file_path):
      print(f"Imagen guardada correctamente en {file_path}")
  else:
      print("No se pudo guardar la imagen.")
```

```python
import matplotlib.pyplot as plt
import cv2

clas_1 = "/content/YOLO-Pre-Trained-Results/Resultado_personas1.png"
clas_2 = "/content/YOLO-Pre-Trained-Results/Resultado_personas5.jpg"
clas_3 = "/content/YOLO-Pre-Trained-Results/Resultado_pose4.jpg"
clas_4 = "/content/YOLO-Pre-Trained-Results/Resultado_bus.jpg"

img_1 = cv2.imread(clas_1)
img_2 = cv2.imread(clas_2)
img_3 = cv2.imread(clas_3)
img_4 = cv2.imread(clas_4)

img_1_color = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)
img_2_color = cv2.cvtColor(img_2, cv2.COLOR_BGR2RGB)
img_3_color = cv2.cvtColor(img_3, cv2.COLOR_BGR2RGB)
img_4_color = cv2.cvtColor(img_4, cv2.COLOR_BGR2RGB)

fig, axes = plt.subplots(1, 4, figsize=(15, 5))

axes[0].imshow(img_1_color)
axes[0].set_title("Personas 1")
axes[0].axis("off")

axes[1].imshow(img_2_color)
axes[1].set_title("Persona")
axes[1].axis("off")

axes[2].imshow(img_3_color)
axes[2].set_title("Mujer")
axes[2].axis("off")

axes[3].imshow(img_4_color)
axes[3].set_title("Bus")
axes[3].axis("off")
```

```python
import matplotlib.pyplot as plt
import cv2
import os

clasificaciones = ["/content/YOLO-Pre-Trained-Results/Resultado_personas1.png",
                  "/content/YOLO-Pre-Trained-Results/Resultado_personas5.jpg",
                  "/content/YOLO-Pre-Trained-Results/Resultado_pose4.jpg",
                  "/content/YOLO-Pre-Trained-Results/pelusa_clas.jpg",
                  "/content/YOLO-Pre-Trained-Results/Resultado_bus.jpg"]

for img_path in clasificaciones:
  img = cv2.imread(img_path)
  img_color = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  plt.imshow(img_color)
  plt.title(os.path.basename(img_path))
  plt.axis("off")
  plt.show()
```
